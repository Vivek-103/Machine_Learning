## Simple Linear Regression

### Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pyplot
```

### Importing the Dataset

```python
dataset = pd.read_csv('Salary_Data.csv')
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

### print x and y

```python
print(x)
```

```python
print(y)
```

### Splitting the Dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=0)
```

### Trainig the Simple Linear Regression model on the Training Set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train , y_train)
```

### Predicting the Test set Result

```python
y_pred = regressor.predict(x_test)
```

### Visualising the Training set Result

```python
pyplot.scatter(x_train , y_train,color='red')
pyplot.plot(x_train ,  regressor.predict(x_train),color='blue')
pyplot.title('Salary vs Experience(training Set)')
pyplot.xlabel('Years of Experience')
pyplot.ylabel('Salary')
pyplot.show
```

### Visualising the Test Set Result

```python
pyplot.scatter(x_test , y_test,color='red')
pyplot.plot(x_train ,  regressor.predict(x_train),color='blue')
pyplot.title('Salary vs Experience(Test Set)')
pyplot.xlabel('Years of Experience')
pyplot.ylabel('Salary')
pyplot.show
```

#### Bonus

##### Making a Single Prediction(salary with 12 years of experience)

```python
print(regressor.predict([[12]]))
```

#### Getting the final linear Equation with Coeffiecients 

```python
print(regressor.coef_)
print(regressor.intercept_)
```
